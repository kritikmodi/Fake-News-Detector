```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Fake-news-detector/Data/Datasets/DataPIB_mini.csv')
```

```python
def get_embed(col) :

  embeddings = []
  a = 1
  
  for i in col :
    print(a)
    embeddings.append(model.encode(i))
    a = a+1

  return embeddings
```

```python
embeddings = get_embed( df['claim'] )
```
