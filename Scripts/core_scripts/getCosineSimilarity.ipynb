```python
embeddings_df = pd.read_csv('/content/drive/MyDrive/SIH/full_embeddings.csv')
```

```python
embeddings_df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'], inplace=True)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
input_text = "the government did not launch financial benefit scheme for farmers"
sen_embeddings = model.encode(input_text)
```

```python
embeddings_df.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

f=0

for i in  range(99146):
    sim = cosine_similarity([sen_embeddings], [embeddings_df.iloc[i]])
    if sim > 0.9:
        f=1
        print('Present')
        break

if not f:
  print('Not Present')
```
