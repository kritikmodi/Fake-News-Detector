```python
def get_embed(col) :

  embeddings = []
  a = 1
  
  for i in col :
    print(a)
    embeddings.append(model.encode(i))
    a = a+1

  return embeddings
```

```python
embeddings = get_embed( df['claim'] )
```

```python
embeddings_df = pd.DataFrame(embeddings)
embeddings_df.to_csv('./DataPIBEmbeddings.csv')
```

```python
inputList = ["Narendra Modi is the PM of India", ]
sen_embeddings = get_embed(inputList)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

f=0

for i in embeddings :
    sim = cosine_similarity(list(i.reshape(1, -1)),list(sen_embeddings))
    if sim > 0.9:
        f=1
        print('Present')
        break

if not f:
  print('Not Present')
```
